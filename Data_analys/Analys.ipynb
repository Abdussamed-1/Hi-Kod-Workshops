```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#Yeni veri seti oluşturalım
df = df[['Gender','Physical_Activity','Previous_Scores','Hours_Studied','Sleep_Hours','Exam_Score','School_Type']]
df.head()
```

```python
#Elimizdeki Yeni Veri Kümesi
df.info()
```

```python
#veri setinin kolonlarına bakalım 
df.columns
```

```python
#Değerlerin Ortalama ve diğer numerik değerlerine bakalım 
df.describe()
```

```python
#Kız Erkek Sayısını öğrenelim
df['Gender'].value_counts().reset_index()
```

```python
#kadın-erkek sayısına histogram grafiği ile bakalım
sns.histplot(x = "Gender", data=df)
```

```python
#Erkek ve Kız bireylerddeki ders çalışma saatlerinin farklılığına bakıyoruz
df[['Gender','Hours_Studied']].value_counts().reset_index()
```

```python
plt.figure(figsize=(8, 6))  # Grafik boyutu
sns.boxplot(x=df['Hours_Studied'])
plt.title("Ders Çalışma Aykırı Değerleri")
plt.show()
```

```python
study = df[['Hours_Studied']]

plt.hist(study, bins=120)
plt.title('Ders Çalışma Dağılım Grafiği')
plt.xlabel("Ders Çalışma Saatleri")
plt.ylabel("Öğrenci Sayıları")
plt.show()
```

```python
#Fiziksel Aktivite Satırında Cinsiyet ders çalışma süresi ve uyku süresine bakalım 
df.groupby('Physical_Activity')[['Gender','Hours_Studied','Sleep_Hours']].value_counts().reset_index()
```

```python
#Ders Çalışma Süresi Fiziksel Aktivite sayısı ve uyku süresinin ortalamalarına bakalım 
df.groupby('Physical_Activity')[['Hours_Studied','Sleep_Hours']].mean().reset_index()
```

```python
#Uyku süresinin Aykırı değerlerine bakıp üstteki tabloyu doğrulayalım
plt.figure(figsize=(8, 6))  # Grafik boyutu
sns.boxplot(x=df['Sleep_Hours'])
plt.title("Uyku Süresi")
plt.show()
```

```python
#Öğrencilerin Uyku sürelerinin sınav skorlarına etkisine bakalım 
df[['Exam_Score','Sleep_Hours']]
```

```python
#Yukarıdaki gösterdiğimiz tablodaki verileri gösterelim
mean_scores = df.groupby('Sleep_Hours')['Exam_Score'].mean().reset_index()

# Çubuk grafiği
plt.bar(mean_scores['Sleep_Hours'], mean_scores['Exam_Score'], color='orange')

# Başlık ve eksenler
plt.title('Ortalama Sınav Skoru ve Uyku Süresi')
plt.xlabel('Uyku Süresi (saat)')
plt.ylabel('Ortalama Sınav Skoru')

# Grafiği göster
plt.show()

```

```python
df[['Previous_Scores','Exam_Score']]
```

```python
# Önceki Sınav Skoru ve Sınav skoru arasındaki koreleasyonu gösterip regresyon çizgisiyle korele olduğunu tespit ettik
sns.jointplot(x='Previous_Scores', y='Exam_Score', data=df, kind='reg',height=15, ratio=7,
              scatter_kws={'color': 'blue'},     # Noktaların rengi
              line_kws={'color': 'red'})         # Regresyon çizgisinin rengi)

# Başlık
plt.suptitle('Sınav Öncesi Skor ve Sınav Skoru Arasındaki Korelasyon', y =1.1)

# Grafiği göster
plt.show()
```

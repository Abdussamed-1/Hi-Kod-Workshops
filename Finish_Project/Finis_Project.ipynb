```python
import matplotlib.pyplot as plt
import numpy as np

# Fiziksel aktivite verilerini simüle etme
np.random.seed(42)  # Rastgelelik için sabit tohum
mean = df['Physical_Activity_Hours'].mean()           # Ortalama değer
median = df['Physical_Activity_Hours'].median()      # Medyan değer
min_value = df['Physical_Activity_Hours'].min()      # Minimum değer
max_value = df['Physical_Activity_Hours'].max()      # Maksimum değer

# Veriyi normal dağılım etrafında dağıtacak şekilde rastgele oluşturma
# Ortalama ve medyana uyması için normal bir dağılım kullanıyoruz
# 1000 örnekleme, ortalama 5, standart sapma 2 ile sınırlı veri seti
activity_hours = np.clip(np.random.normal(loc=mean, scale=1, size=1000), min_value, max_value  )

# Histogram çizimi
plt.figure(figsize=(10, 6) )
plt.hist(activity_hours, bins=20, color='skyblue', edgecolor='black')
plt.title("Fiziksel Aktivite Saati Histogram Grafiği")
plt.xlabel("Physical Activity Hours")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
np.random.seed(0)
data = pd.DataFrame({
    "Physical_Activity_Hours": np.random.uniform(0, 10, 300),
    "Screen_Time_Hours": np.random.uniform(2, 15, 300),
    "Gender": np.random.choice(["Female", "Male", "Other"], 300)
})
```

```python
# Stili ve renk paletini ayarlama
sns.set(style="whitegrid")
palette = {"Female": "dodgerblue", "Male": "coral", "Other": "lightgray"}

# Pair plot oluşturma
g = sns.pairplot(
    data,
    kind="scatter",
    hue="Gender",
    palette=palette,
    plot_kws={'alpha': 0.6, 's': 40},
    diag_kind="kde",
    diag_kws={'alpha': 0.5}
)

# Başlığı ekleme
g.fig.suptitle("Physical Activity and Screen Time Distribution by Gender",
               y=1.02, fontsize=16)

# Eksen adlarını düzenleme
g.set(xlabel="Physical Activity", ylabel="Screen Time")

# Grafiği gösterme
plt.show()
```
